```python
#check if a number is even or odd
num = int(input("Enter a number:  " ))
if num % 2 == 0:
  print("number : even")
else:
  print("number: odd")
```

```python
# check if a number is divisible by 5
number = int(input("Enter a number:  "))
if number % 5 ==0:
  print("Number is divisible by 5")
else:
  print("Number is not divisible by 5")
```

```python
# check if the number is divisible by 3 & 5
num = int(input("Enter a number"))
if num % 5 ==0 & num % 3 ==0:
  print("Number is divisible by 3 & 5")
else:
  print("Number is not divisible by 3 & 5 ")
```

```python
#check year is leap year or not
year = int(input("Enter a year:"))
if year % 4 == 0:
  print("Leap year")
else:
  print("Not leap year")
```

```python
#Take a temperature value and print "cold", "warm" and 'Hot using range conditions
temper = int(input("Enter temperature value: "))
if temper <= 10:
 print("Cold")
elif temper <= 20:
 print("Warm")
elif temper <= 30:
 print("Hot")
else:
  print("very hot")
```

```python
#Take a character and check if it's a vowel or consonant
letter = input("Enter a character: ").lower()   # convert to lower case
if letter in 'a,e,i,o,u':
 print("vowel")
else:
 print("consonant")
```

```python
#Take a character and check whether it's uppercase, lowercase , a digit, or a special character
char = input("Enter a character: ")
if char.isupper():
  print("uppercase")
elif char.islower():
  print("lowercase")
elif char.isdigit():
  print("digits")
else:
  print("special characters")

```
