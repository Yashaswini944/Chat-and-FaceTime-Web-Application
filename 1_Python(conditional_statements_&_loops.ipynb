```python
#checking a condition
num = 2
if num > 0:
  print("positive Number")
elif num == 0:
  print('Zero')
else:
  print("Negative number")
```

```python
#For Loop through a List
fruits =['apple','mango','cherry']
for n in fruits:
  print(n)
```

```python
#while loop
count = 3
while count > 0:
  print(count)
  count -= 1
print("outside while loop")
```

```python
#using break and continue
for i in range(10):
  if i == 4:
    break
  print(i)
print("Out of")

#odd number
for i in range(10):
  if i % 2 ==0:
    continue
  print(i)


```

```python
#Exercise 1:create a Menu-Driven Calculator
def add(a, b):
    return a + b

def subtract(a, b):
    return a - b

def multiply(a, b):
    return a * b

def divide(a, b):
    if b == 0:
        return "Error! Division by zero."
    return a / b


while True:
    print("\n--- Menu-Driven Calculator ---")
    print("1. Addition")
    print("2. Subtraction")
    print("3. Multiplication")
    print("4. Division")
    print("5. Exit")

    choice = input("Enter your choice (1-5): ")

    if choice == "5":
        print("Calculator closed. Bye ðŸ‘‹")
        break

    if choice in ["1", "2", "3", "4"]:
        num1 = float(input("Enter first number: "))
        num2 = float(input("Enter second number: "))

        if choice == "1":
            print("Result:", add(num1, num2))
        elif choice == "2":
            print("Result:", subtract(num1, num2))
        elif choice == "3":
            print("Result:", multiply(num1, num2))
        elif choice == "4":
            print("Result:", divide(num1, num2))
    else:
        print("Invalid choice! Please select 1 to 5.")



```

```python
#write a program to calculate the factorial of a number using a while loop

num = int(input("Enter a non-negative integer: "))

# Check if number is negative
if num < 0:
    print("Factorial is not defined for negative numbers.")
else:
    factorial = 1
    i = 1

    while i <= num:
        factorial *= i
        i += 1

    print(f"Factorial of {num} is {factorial}")
```

```python
#Create a program to find the largest number in a list using a for loop

numbers = [10, 45, 23, 67, 12, 89, 34]

# Assume first element is the largest
largest = numbers[0]

for num in numbers:
    if num > largest:
        largest = num

print("The largest number in the list is:", largest)
```
